```python
# Python code to convert string to list

def convert(string):
    li = list(string.split(","))
    return li
```

```python
f1 = open("results/datafiles/caesarfile_snapshot_150_data.txt","r")
```

```python
x = f1.readline()
print(len(x),x[1:-2])
```

```python
stellar_masses_str = convert(x[1:-2])
stellar_masses = []

for i in range(0,len(stellar_masses_str)):
    stellar_masses.append(float(stellar_masses_str[i]))
```

```python
gas_masses_str = convert(x[1:-2])
gas_masses = []

for i in range(0,len(gas_masses_str)):
    gas_masses.append(float(gas_masses_str[i]))
```

```python
sfr_str = convert(x[1:-2])
sfr = []

for i in range(0,len(sfr_str)):
    sfr.append(float(sfr_str[i]))
```

```python
specific_sfr = []

for i in range(0,len(sfr)):
    specific_sfr.append(sfr[i]/stellar_masses[i])
```

```python
mass_weighted_gas_Z_str = convert(x[1:-2])
mass_weighted_gas_Z = []

for i in range(0,len(mass_weighted_gas_Z_str)):
    mass_weighted_gas_Z.append(float(mass_weighted_gas_Z_str[i]))
```

```python
f1.close()
```

```python
plt.rc('axes', labelsize=13)
plt.scatter(np.log10(stellar_masses),np.log10(specific_sfr),s=0.1,color="blue")
#plt.xscale('log')
#plt.yscale('log')
plt.xlabel("log $M_{*}$")
plt.ylabel("log sSFR")
plt.show()
```

```python
plt.rc('axes', labelsize=13)
plt.scatter(np.log10(stellar_masses),np.log10(sfr),s=0.1,color="blue")
plt.scatter(np.log10(stellar_masses[391]),np.log10(sfr[391]),s=5,color="red")
plt.xlabel("log $M_{*}$")
plt.ylabel("log SFR")
plt.show()
```
