```python
densities = np.array([0.1, 1, 100, 1000])
SFRs = np.array([0.01, 1, 10, 100])
metals = np.array([0.001,0.01,0.1,1])
temps = np.array([1e4,1e5,1e6,1e7])
```

```python
grid = np.zeros((256,7))

i=0

kB = 1.3807 * 10**(-16)

for dens in range(0,len(densities)):
    for sfr in range(0,len(SFRs)):
        for metal in range(0,len(metals)):
            for temp in range(0,len(temps)):
                grid[i][0] = densities[dens]
                grid[i][1] = SFRs[sfr]
                grid[i][2] = metals[metal]
                grid[i][3] = temps[temp]
                grid[i][4] = 1e6     #mass of cloud
                grid[i][5] = 2.2620999761958926    #redshift
                P = densities[dens]*kB*temps[temp]
                grid[i][6] =  (P/(kB*1.e4))**(-0.25) * (1e6/(290))**0.5 #radius of cloud
                i=i+1
```

```python
index = np.linspace(0,255,256)
df = pd.DataFrame({"c_nDensity":grid[:,0],"g_SFR":grid[:,1],"c_Metallicity":grid[:,2],"c_Mass":grid[:,4],"g_Redshift":grid[:,5],"c_Radius":grid[:,6],"c_Index":index})
```

```python
coords = pd.read_csv("results/datafiles/coords_750.csv")

def distance(x1,y1,z1,x2,y2,z2):
    
    val = np.sqrt((x1-x2)**2+(y1-y2)**2+(z1-z2)**2)
    return val

min_dist = 1000

for i in range(0,len(coords)):
    
    for j in range(0,len(coords)):
        
        if i!=j:
            
            dist = distance(coords['x'][i],coords['y'][i],coords['z'][i],coords['x'][j],coords['y'][j],coords['z'][j])
        
        if dist<min_dist:
            min_dist = dist
            print(coords['x'][i],coords['y'][i],coords['z'][i],coords['x'][j],coords['y'][j],coords['z'][j],min_dist)
```
